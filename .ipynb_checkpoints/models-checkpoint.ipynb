```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor,
    ExtraTreesRegressor,
    AdaBoostRegressor,
    VotingRegressor
)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
# Event signals
df['views'] = (df['event_type'] == 'view').astype(int)
df['purchases'] = (df['event_type'] == 'purchase').astype(int)
df['cart'] = (df['event_type'] == 'cart').astype(int)

# Aggregate per product
agg_df = df.groupby('product_id').agg({
    'views': 'sum',
    'purchases': 'sum',
    'cart': 'sum',
    'price': 'mean'
}).reset_index()

# Demand
agg_df['demand'] = agg_df['views'] + agg_df['cart'] + agg_df['purchases']

# Conversion rate (NEW 🔥)
agg_df['conversion_rate'] = agg_df['purchases'] / (agg_df['views'] + 1)

# Cart ratio (NEW 🔥)
agg_df['cart_ratio'] = agg_df['cart'] / (agg_df['views'] + 1)

# Log transform demand (FIX scaling issue 🔥)
agg_df['log_demand'] = np.log1p(agg_df['demand'])

agg_df.head()
```

```python
# Instead of random multiplier → use realistic signals
agg_df['target_price'] = (
    agg_df['price'] * 
    (1 + 0.3 * agg_df['conversion_rate'] + 0.2 * agg_df['cart_ratio'])
)

# Log transform target (IMPORTANT 🔥)
agg_df['target_price'] = np.log1p(agg_df['target_price'])

agg_df[['price', 'target_price']].head()
```

```python
features = [
    'views',
    'purchases',
    'cart',
    'log_demand',
    'conversion_rate',
    'cart_ratio'
]

X = agg_df[features]
y = agg_df['target_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Extra Trees": ExtraTreesRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "AdaBoost": AdaBoostRegressor()
}
```

```python
def evaluate(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return mae, mse, rmse, r2
```

```python
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    results[name] = evaluate(y_test, preds)
```

```python
ensemble = VotingRegressor([
    ('gb', GradientBoostingRegressor()),
    ('rf', RandomForestRegressor())
])

ensemble.fit(X_train, y_train)
ensemble_preds = ensemble.predict(X_test)

results['Ensemble'] = evaluate(y_test, ensemble_preds)
```

```python
results_df = pd.DataFrame(results, index=['MAE','MSE','RMSE','R2']).T
results_df
```

```python
results_df[['MAE','RMSE']].plot(kind='bar', figsize=(10,5))
plt.title("Error Comparison")
plt.show()

results_df[['R2']].plot(kind='bar', figsize=(10,5))
plt.title("R2 Comparison")
plt.show()
```

```python
best_model_name = results_df['MAE'].idxmin()

if best_model_name == "Ensemble":
    best_model = ensemble
else:
    best_model = models[best_model_name]

print("Best Model:", best_model_name)
```

```python
def dynamic_price(product_id, user_type="low_spender"):
    
    row = agg_df[agg_df['product_id'] == product_id].iloc[0]
    
    input_data = np.array([[
        row['views'],
        row['purchases'],
        row['cart'],
        row['log_demand'],
        row['conversion_rate'],
        row['cart_ratio']
    ]])
    
    input_scaled = scaler.transform(input_data)
    
    base_price = row['price']
    
    # ML prediction
    ml_price = np.expm1(best_model.predict(input_scaled))[0]
    
    # RULE-BASED CORE (IMPORTANT 🔥)
    price = base_price
    
    if row['demand'] > 50:
        price *= 1.1
        
    elif row['demand'] < 10:
        price *= 0.9
    
    # ML adjustment
    price = 0.7 * price + 0.3 * ml_price
    
    # Competitor
    competitor_price = base_price * 0.95
    if competitor_price < price:
        price *= 0.95
    
    # User
    if user_type == "low_spender":
        price *= 0.9
    
    return round(price, 2)
```

```python
sample_product = agg_df['product_id'].iloc[0]

print("Final Price:", dynamic_price(sample_product))
```

```python
# Get some product IDs
agg_df['product_id'].head(5)
product_id = agg_df['product_id'].iloc[0]

price = dynamic_price(product_id, user_type="low_spender")

print("Product ID:", product_id)
print("Predicted Price:", price)
```
